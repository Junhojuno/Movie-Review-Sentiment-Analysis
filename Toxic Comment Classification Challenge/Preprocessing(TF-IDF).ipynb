```python
import re
def clean_comment(text):
    text = text.lower()
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "cannot ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text)
    text = re.sub('mother fucking', 'motherfucking', text)
    text = re.sub('cock sucker', 'cocksucker', text)
    text = re.sub('son bitch', 'sonofbitch', text)
    
    text = text.strip(' ')
    
    return text
```

```python
train['comment_text'] = train['comment_text'].map(lambda x: clean_comment(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
cnt_vect = CountVectorizer(min_df=2  ,stop_words='english')
```

```python
toxic = train[train['toxic'] == 1]
toxic_cnt_vector = cnt_vect.fit(toxic.comment_text)
```
